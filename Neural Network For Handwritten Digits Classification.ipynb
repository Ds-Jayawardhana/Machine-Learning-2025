```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test) =keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattend=X_train.reshape(len(X_train),28*28)
X_test_flattend=X_test.reshape(len(X_test),28*28)

```

```python
X_test_flattend.shape
```

```python
X_test_flattend[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattend, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattend,y_test)
```

```python
plt.matshow(X_test[1])
plt.show()
```

```python
y_predicted=model.predict(X_test_flattend)
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show()
```

Adding a Hidden Layer to the Previous Model

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattend, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattend,y_test)
```

```python
y_predicted=model.predict(X_test_flattend)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show()
```

A Chat Gpt code to Increase Accuracy up to 100%

```python
model = keras.Sequential([
    keras.layers.Dense(512, input_shape=(784,), activation='relu'),  # First hidden layer with 512 neurons
    keras.layers.Dense(256, activation='relu'),  # Second hidden layer with 256 neurons
    keras.layers.Dense(128, activation='relu'),  # Third hidden layer with 128 neurons
    keras.layers.Dense(64, activation='relu'),   # Fourth hidden layer with 64 neurons
    keras.layers.Dense(10, activation='softmax')  # Output layer for 10 classes (digits 0-9)
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattend, y_train, epochs=10, validation_split=0.1)  # Increased epochs and added validation
```

```python
y_predicted=model.predict(X_test_flattend)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show()
```
